# water jug problem


```python
from collections import deque

def bfs(start, end, x_capacity, y_capacity):
    path = []
    front = deque()
    front.append(start)
    visited = set()  # Using a set for faster lookup
    while front:
        current = front.popleft()
        x, y = current[0], current[1]
        path.append(current)
        if x == end or y == end:
            print("Found")
            return path
        # rule 1
        if x < x_capacity and (x_capacity, y) not in visited:
            front.append((x_capacity, y))
            visited.add((x_capacity, y))
        # rule 2
        if x > 0 and (0, y) not in visited:
            front.append((0, y))
            visited.add((0, y))
        # rule 3
        if y < y_capacity and (x, y_capacity) not in visited:
            front.append((x, y_capacity))
            visited.add((x, y_capacity))
        # rule 4
        if y > 0 and (x, 0) not in visited:
            front.append((x, 0))
            visited.add((x, 0))
        # rule 5
        if x + y <= x_capacity and y > 0 and (x + y, 0) not in visited:
            front.append((x + y, 0))
            visited.add((x + y, 0))
        # rule 6
        if x + y <= y_capacity and x > 0 and (0, x + y) not in visited:
            front.append((0, x + y))
            visited.add((0, x + y))
        # rule 7
        if x + y >= x_capacity and y > 0 and (x_capacity, y - (x_capacity - x)) not in visited:
            front.append((x_capacity, y - (x_capacity - x)))
            visited.add((x_capacity, y - (x_capacity - x)))
        # rule 8
        if x + y >= y_capacity and x > 0 and (x - (y_capacity - y), y_capacity) not in visited:
            front.append((x - (y_capacity - y), y_capacity))
            visited.add((x - (y_capacity - y), y_capacity))
    return "Not found"

def gcd(a, b):
    if a == 0:
        return b
    return gcd(b % a, a)

x_capacity = int(input("Enter Jug 1 capacity: "))
y_capacity = int(input("Enter Jug 2 capacity: "))
end = int(input("Enter target volume: "))

start = (0, 0)
if end % gcd(x_capacity, y_capacity) == 0:
    print(bfs(start, end, x_capacity, y_capacity))
else:
    print("No solution possible for this combination.")
```

```python
def dfs(start, end, x_capacity, y_capacity, visited):
    if start in visited:
        return None

    visited.add(start)
    x, y = start

    if x == end or y == end:
        return [start]

    # Rule 1
    if x < x_capacity:
        path = dfs((x_capacity, y), end, x_capacity, y_capacity, visited)
        if path:
            return [start] + path

    # Rule 2
    if x > 0:
        path = dfs((0, y), end, x_capacity, y_capacity, visited)
        if path:
            return [start] + path

    # Rule 3
    if y < y_capacity:
        path = dfs((x, y_capacity), end, x_capacity, y_capacity, visited)
        if path:
            return [start] + path

    # Rule 4
    if y > 0:
        path = dfs((x, 0), end, x_capacity, y_capacity, visited)
        if path:
            return [start] + path

    # Rule 5
    if x + y <= x_capacity:
        path = dfs((x + y, 0), end, x_capacity, y_capacity, visited)
        if path:
            return [start] + path

    # Rule 6
    if x + y <= y_capacity:
        path = dfs((0, x + y), end, x_capacity, y_capacity, visited)
        if path:
            return [start] + path

    # Rule 7
    if x + y >= x_capacity:
        path = dfs((x_capacity, y - (x_capacity - x)), end, x_capacity, y_capacity, visited)
        if path:
            return [start] + path

    # Rule 8
    if x + y >= y_capacity:
        path = dfs((x - (y_capacity - y), y_capacity), end, x_capacity, y_capacity, visited)
        if path:
            return [start] + path

    return None

def water_jug_dfs(x_capacity, y_capacity, end):
    start = (0, 0)
    visited = set()
    path = dfs(start, end, x_capacity, y_capacity, visited)
    if path:
        print("Found")
        return path
    else:
        print("No solution possible for this combination.")
        return []

x_capacity = int(input("Enter Jug 1 capacity: "))
y_capacity = int(input("Enter Jug 2 capacity: "))
end = int(input("Enter target volume: "))

print(water_jug_dfs(x_capacity, y_capacity, end))
```
